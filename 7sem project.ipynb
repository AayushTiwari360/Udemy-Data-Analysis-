#### 7. Shows the trend of course publication over time, In this you see whether Udemy’s content creation has grown.

```python

df['published_timestamp'] = pd.to_datetime(df['published_timestamp'], errors='coerce')


df['published_year'] = df['published_timestamp'].dt.year


yearly_courses = df['published_year'].value_counts().sort_index()


plt.figure(figsize=(10,6))
sns.lineplot(x=yearly_courses.index, y=yearly_courses.values, marker='o', color='teal')
plt.title('Number of Courses Published Each Year')
plt.xlabel('Year')
plt.ylabel('Number of Courses')
plt.grid(True)
plt.show()
```

## Apllying Clustering Machine Learning ALgorithms 

####  Clustering 

In clustering , we use only numeric columns .   
Using Sklearn Library for cluster analysis.

```python

df['price'] = df['price'].replace('Free', 0)  #---> Course ka price agr free ho tho , 0 put kro 


df['price'] = pd.to_numeric(df['price'], errors='coerce')


features = ['num_subscribers', 'num_reviews', 'num_lectures', 'price']
df_cluster = df[features].dropna()


from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

scaler = StandardScaler()
scaled_data = scaler.fit_transform(df_cluster)

# Elbow method to find best k
inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_data)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(range(1, 11), inertia, marker='o')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()

# K means clustering algorithm
kmeans = KMeans(n_clusters=3, random_state=42)
df_cluster['Cluster'] = kmeans.fit_predict(scaled_data)


plt.figure(figsize=(8, 6))
sns.scatterplot(
    x=df_cluster['num_subscribers'],
    y=df_cluster['price'],
    hue=df_cluster['Cluster'],
    palette='viridis'
)
plt.title('Clusters of Udemy Courses')
plt.show()
```
